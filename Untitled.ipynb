```python
# We'll need numpy for some mathematical operations
import numpy as np

# matplotlib for displaying the output
import matplotlib.pyplot as plt
import matplotlib.style as ms
ms.use('seaborn-muted')
%matplotlib inline

# and IPython.display for audio output
import IPython.display

# Librosa for audio
import librosa
# And the display module for visualization
import librosa.display

from moviepy.editor import *
from moviepy.video.VideoClip import ImageClip
from random import choice
```

```python
class Size(object):
    def __init__(self, height, width):
        self.width = width
        self.height = height

class Settings(object):
    def __init__(self, audio_path, video_output_path, grid_size, video_size):
        self.audio_path = audio_path
        self.video_output_path = video_output_path
        self.grid_size = grid_size
        self.video_size = video_size
        
def work(s):
    y, sr = librosa.load(s.audio_path)
    
    S = librosa.feature.melspectrogram(y, sr=sr, n_mels=128)

    log_S = librosa.power_to_db(S, ref=np.max)


    plt.figure(figsize=(12,4))

    librosa.display.specshow(log_S, sr=sr, x_axis='time', y_axis='mel')

    plt.title('mel power spectrogram')

    plt.colorbar(format='%+02.0f dB')

    plt.tight_layout()
    
    y_harmonic, y_percussive = librosa.effects.hpss(y)
    
    plt.figure(figsize=(12, 6))
    onset_env = librosa.onset.onset_strength(y, sr=sr) #,
#                                           aggregate=np.median)
#     tempo, raw_beats = librosa.beat.beat_track(onset_envelope=onset_env, sr=sr)
    hop_length = 512
    times = librosa.frames_to_time(np.arange(len(onset_env)),
                                sr=sr, hop_length=hop_length)
#     beats = []
#     for x in raw_beats:
#         onset = librosa.util.normalize(onset_env)[x]
#         if onset > 0.2:
#             beats.append(times[x])

    beats = []
    for i, x in enumerate(librosa.util.normalize(onset_env)):
        if x > 0.15:
            beats.append(times[i])

#     print('beats size is: %d' % len(beats))
    
    w = int(s.video_size.width / s.grid_size.width)
    h = int(s.video_size.height / s.grid_size.height)
        
    
    def generate_image(color=(0, 0, 0)):
        arr = np.zeros([h, w, 3])
        for x in range(h):
            for y in range(w):
                arr[x][y] = np.array(color) * 255
        return arr
    
    template_clips = [
#         ImageClip(generate_image()), 
        ImageClip(generate_image((1, 0, 0))),
        ImageClip(generate_image((0, 1, 0))),
        ImageClip(generate_image((0, 0, 1))),
        ImageClip(generate_image((1, 1, 0))),
        ImageClip(generate_image((1, 1, 1))),
        ImageClip(generate_image((1, 0, 1))),
#         ImageClip(generate_image((1, 1, 1))),
    ]
    
#     audio_clip = AudioFileClip('/Users/daniiltulin/sample_song.mp3')

    clips_by_rect = []
    for x in range(s.grid_size.width):
        clips_by_rect.append([])
        for y in range(s.grid_size.height):
            clips_by_rect[x].append([])
                
    for i in range(-1, len(beats)):
        if i == len(beats) - 1:
            continue
        if i == -1:
            start = 0
        else:
            start = beats[i]
        end = beats[i + 1]
    
        frame_clips = []
        
        for x in range(s.grid_size.width):
            for y in range(s.grid_size.height):                
                clip = choice(template_clips).copy(). \
                    set_start(start). \
                    set_end(end)
                clips_by_rect[x][y].append(clip)
    
    concatenated_clips = []
    for x in range(s.grid_size.width):
            for y in range(s.grid_size.height):
                x_offset = float(x) / s.grid_size.width
                y_offset = float(y) / s.grid_size.height
                concatenated_clip = concatenate_videoclips(clips_by_rect[x][y]). \
                    set_position((x_offset, y_offset), relative=True)
                print('position is %f %f' % (x_offset, y_offset))
                concatenated_clips.append(concatenated_clip)
    final = CompositeVideoClip(concatenated_clips)
#     final.audio = AudioFileClip(s.audio_path)
    final.write_videofile(
        s.video_output_path,codec='libx264',fps=24,threads=12,
        audio=s.audio_path,
    )
```

```python
s = Settings(
    audio_path='/Users/daniiltulin/Downloads/Leaf - Nada [OFFICIAL VIDEO] ft. Lil Yachty.mp3',
    video_output_path='foo2.avi',
    grid_size=Size(height=3, width=4),
    video_size=Size(height=720, width=1280),
)
work(s)
```

```python
y, sr = librosa.load('/Users/daniiltulin/Downloads/The Black Eyed Peas - Lets Get It Started.mp3')
onset_env = librosa.onset.onset_strength(y, sr=sr)#,
#                                           aggregate=np.median)
tempo, beats = librosa.beat.beat_track(onset_envelope=onset_env,
                                       sr=sr)
```

```python
hop_length = 512
plt.figure(figsize=(8, 4))
times = librosa.frames_to_time(np.arange(len(onset_env)),
                                sr=sr, hop_length=hop_length)
plt.plot(times, librosa.util.normalize(onset_env),
         label='Onset strength')
plt.vlines(times[beats], 0, 1, alpha=0.5, color='r',
           linestyle='--', label='Beats')
plt.legend(frameon=True, framealpha=0.75)
# Limit the plot to a 15-second window
plt.xlim(50, 60)
plt.gca().xaxis.set_major_formatter(librosa.display.TimeFormatter())
plt.tight_layout()
```
